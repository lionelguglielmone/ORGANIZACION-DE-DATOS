```python
import pandas as pd
import numpy as np
```

```python
text_numerical_features = pd.read_csv('text_numerical_features.csv')
text_numerical_features.iloc[:,0:20]
```

```python
text_numerical_features_palabras_y_caracteres = text_numerical_features[['target','#caracteres','#palabras','#palabras_unicas','#stopwords']]
text_numerical_features_palabras_y_caracteres
```

```python
text_numerical_features_palabras_y_caracteres_aux = text_numerical_features_palabras_y_caracteres.groupby('target').agg({'#caracteres':'mean','#palabras':'mean','#palabras_unicas':'mean','#stopwords':'mean'})
text_numerical_features_palabras_y_caracteres_aux = text_numerical_features_palabras_y_caracteres_aux.reset_index()
text_numerical_features_palabras_y_caracteres_aux
```

```python
text_numerical_features_palabras_y_caracteres_aux.plot(x="target", y=["#caracteres", "#palabras", "#palabras_unicas",'#stopwords'], kind="bar",title='VALORES PROMEDIO VARIABLES GENERALES', figsize = [12,6])
```

```python
puntuacion_text_numerical_features = text_numerical_features[['target','#puntuacion','#capitalize','#mayusculas','#caracteres_especiales']]
puntuacion_text_numerical_features
```

```python
puntuacion_text_numerical_features_aux = puntuacion_text_numerical_features.groupby('target').agg({'#puntuacion':'mean','#capitalize':'mean','#mayusculas':'mean','#caracteres_especiales':'mean'})
puntuacion_text_numerical_features_aux = puntuacion_text_numerical_features_aux.reset_index()
puntuacion_text_numerical_features_aux
```

```python
puntuacion_text_numerical_features_aux.plot(x="target", y=["#puntuacion", "#capitalize", "#mayusculas",'#caracteres_especiales'], kind="bar",title='VALORES PROMEDIO PUNTUACION', figsize = [12,6])
```
