```python
import pandas as pd
import numpy as np
```

```python
train_location_numerical = pd.read_csv('train_location_numerical.csv')
train_location_numerical
```

```python
train_location_numerical = train_location_numerical.groupby('target').agg({'has_country':'sum', 'has_city':'sum','has_state':'sum', 'has_county':'sum', 'has_continent':'sum'})
train_location_numerical
```

```python
train_location_numerical.plot(kind = 'bar', figsize = [12,9])
```
