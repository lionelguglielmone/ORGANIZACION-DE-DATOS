### 2) Dados los archivos: - Notas (Padron, Codigo de Materia, Codigo de Curso, Nota) - Cursos (Codigo de Materia, Nombre de Materia, Codigo de Curso, - Profesor a Cargo)
### Hacer un programa en Pandas que liste para cada curso de cada materia, el promedio de notas de los alumnos que aprobaron la misma. El listado debe contener Codigo de Materia, Codigo de Curso, Profesor a cargo y Promedio de Notas, y debe estar ordenado por materia. Tener en cuenta solo los cursos que tengan al menos 5 alumnos aprobados. aero. (***) (15 pts)

```python
notas = pd.read_csv('notas.csv')
notas.head()
```

```python
cursos = pd.read_csv('cursos.csv')
cursos.head()
```

```python
aprobados = notas[notas['nota']>3]
aprobados = aprobados [['codigo_de_materia','codigo_curso','nota']]
aprobados ['contador'] = 1
aprobados.head()
```

```python
aprobados_por_materia_y_curso  = aprobados.groupby(['codigo_de_materia','codigo_curso']).agg({'nota':'mean','contador':'count'})
aprobados_por_materia_y_curso.head(10)
```

```python
aprobados_por_materia_y_curso = aprobados_por_materia_y_curso.reset_index()
aprobados_por_materia_y_curso = aprobados_por_materia_y_curso[aprobados_por_materia_y_curso['contador']>4]
aprobados_por_materia_y_curso.head()
```

```python
listado_aprobados_destacable = pd.merge(aprobados_por_materia_y_curso, cursos, on = ['codigo_de_materia','codigo_curso'])
listado_aprobados_destacable = listado_aprobados_destacable[['codigo_de_materia','codigo_curso','profesor','nota']]
listado_aprobados_destacable
```
