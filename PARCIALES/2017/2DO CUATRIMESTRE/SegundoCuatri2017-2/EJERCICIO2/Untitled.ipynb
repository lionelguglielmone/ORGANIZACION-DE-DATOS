```python
import pandas as pd
import numpy as np
```

### 2) La Agencia Nacional de Estadísticas de Buenos Aires recolecta información de nacimientos cuando los padres registran a sus hijos en el registro civil a partir de una encuesta. Esa información se encuentra disponible para su análisis en un csv con el siguiente formato (dia_nacimiento, mes_nacimiento, anio_nacimiento, peso_al_nacer, longitud_al_nacer, id_hospital, tipo_parto), donde el tipo de parto 1 es natural y 2 es cesárea. Por  otro lado la agencia cuenta con información histórica de los hospitales en otro csv con siguiente formato id_hospital, dirección, promedio_nacimientos_mensual).
### Se pide usar Pandas para:
### a) Calcular la cantidad de nacimientos para cada uno de los hospitales para el mes de Octubre de 2017 e indicar aquellos hospitales que superan el promedio de nacimientos mensuales.
### b) Comparando el mes de Octubre de 2017 indicar programáticamente si se incremento el % de cesáreas con respecto a ese mes del año 2016. (****) (15 pts) 


```python
nacimientos = pd.read_csv('nacimientos.csv')
nacimientos
```

```python
promedio_octubre2016= pd.read_csv('promedio_nacimientos_octubre2016.csv')
promedio_octubre2016
```

```python
# a)Calcular la cantidad de nacimientos para cada uno de los hospitales para el mes de Octubre de 2017 e indicar aquellos 
# hospitales que superan el promedio de nacimientos mensuales.
nacimientos_octubre2017 = (nacimientos[(nacimientos['mes_nacimiento']==10) & (nacimientos['anio_nacimiento']==2017)])
nacimientos_octubre2017 ['cantidad'] = 1
nacimientos_octubre2017_short = nacimientos_octubre2017[['id_hospital','cantidad']]
nacimientos_octubre2017_short.head(10)
```

```python
nacimientos_octubre2017_aux= nacimientos_octubre2017_short.groupby('id_hospital', as_index=False).agg('count')
nacimientos_octubre2017_aux
```

```python
#calculo el promedio de nacimientos para Octubre 2017
nacimientos_octubre2017_aux ['promedio_octubre2017'] = nacimientos_octubre2017_aux['cantidad'].mean()
nacimientos_octubre2017_aux['supera_nacimiento_promedio'] = nacimientos_octubre2017_aux ['cantidad']>nacimientos_octubre2017_aux ['promedio_octubre2017']
nacimientos_octubre2017_aux
```

```python
# b) Comparando el mes de Octubre de 2017 indicar programáticamente si se incremento el % de cesáreas con 
# respecto a ese mes del año 2016. (****) (15 pts)
nacimientos_septiembre2017 = (nacimientos[(nacimientos['mes_nacimiento']==9) & (nacimientos['anio_nacimiento']==2017) & (nacimientos['tipo_parto']==2)])
nacimientos_septiembre2017_short = nacimientos_septiembre2017[['mes_nacimiento']]
nacimientos_septiembre2017_short['cantidad_cesareas_septiembre2017'] = 1
nacimientos_septiembre2017_short.head()
```

```python
nacimientos_septiembre2017_cesarea = nacimientos_septiembre2017_short.groupby('cantidad_cesareas_septiembre2017').agg('count')
nacimientos_septiembre2017_cesarea
```

```python
nacimientos_octubre2017 = (nacimientos[(nacimientos['mes_nacimiento']==10) & (nacimientos['anio_nacimiento']==2017) & (nacimientos['tipo_parto']==2)])
nacimientos_octubre2017_short = nacimientos_octubre2017[['mes_nacimiento']]
nacimientos_octubre2017_short['cantidad_cesareas_octubre2017'] = 1
nacimientos_octubre2017_short.head()
```

```python
nacimientos_octubre2017_cesarea = nacimientos_octubre2017_short.groupby('cantidad_cesareas_octubre2017').agg('count')
nacimientos_octubre2017_cesarea
```

```python
variacion_porcentual_cesareas_septiembre_octubre = (nacimientos_octubre2017_cesarea.iloc[0,0]*100/nacimientos_septiembre2017_cesarea.iloc[0,0])-100
print('La variacion porcentual es de: ', variacion_porcentual_cesareas_septiembre_octubre)
```
