### 2) Tenemos un dataframe con la informaci√≥n de distintas playlists armadas por usuarios con el formato (playlist, song_id, description). A su vez, contamos con un dataframe de canciones que contiene (song_id, singer, year, lenght, genres).  Se pide generar un programa en Pandas que indique para cada playlist cual es el cantante predominante (con mas canciones incluidas dentro de esa lista). (**) (15 pts) 

```python
playlists = pd.read_csv('playlists.csv')
playlists_short = playlists [['playlist','song_id']]
playlists_short.head()
```

```python
canciones = pd.read_csv ('canciones.csv')
canciones_short = canciones[['song_id','singer']]
canciones_short.head()
```

```python
#Unifico ambos DF's y agrego columna cant para contabilizar ocurrencias despues
playlist_canciones = pd.merge(canciones_short,playlists_short, on ='song_id')
playlist_canciones['cant'] = 1
playlist_canciones.head()
```

```python
#Me quedo con las columnas de utilidad - no es necesario este paso
playlist_and_singers = playlist_canciones[['playlist','singer','cant']]
playlist_and_singers.head()
```

```python
#Agrupo por columnas 'playlist' y 'singer' y contabilizo ocurrencias del cantante para ESA 'playlist'
singers_por_playlist = playlist_canciones.groupby(['playlist','singer'])['cant'].agg('count')
singers_por_playlist.head()
```

```python
#Reseteo index
singers_por_playlist = singers_por_playlist.reset_index()
singers_por_playlist.head()
```

```python
#Agrupo por playlist y guardo el valor maximo para cada playlist en una nueva columna
singers_por_playlist['count_max'] = singers_por_playlist.groupby(['playlist'])['cant'].transform(max)
singers_por_playlist.head()
```

```python
#Para quedarme con los mas escuchados por playlist, pido igualdad entre valores de columnas 'cant' y 'count_max'
mas_escuchados_por_playlist = singers_por_playlist[singers_por_playlist['count_max']==singers_por_playlist['cant']]
```

```python
mas_escuchados_por_playlist = mas_escuchados_por_playlist[['playlist','singer','cant']]
mas_escuchados_por_playlist.head(20)
```
