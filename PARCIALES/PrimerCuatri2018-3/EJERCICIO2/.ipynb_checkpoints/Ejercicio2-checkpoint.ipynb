## 2)El equipo comercial en Argentina de instagram cuenta con informacion sumarizada de usuarios del pais en el siguiente Dta Frame (id_usuario, nickname, pais, fecha_alta, cant_seguidores, cant_seguidos, cant_tot_publicaciones). Uno de sus analistas quiere obtener los potenciales 10 influencers. El criterio para detectar un influencer es que el mismo debe:
### -Tener una cant de seguidores mayor al 80% de la maxima cant de seguidores de todos los usuarios dentro de su pais (Por ej: si un usuario es Argentino y el  max de seguidores en Argentina es 1000 entonces requerira tener al menos 800 seguidores).
### -El nuemro de usuarios que esa persona sigue no debe superar el 15% de la cant de seguidores que el tiene.
##### El criterio final para seleccionar a los candidatos sera a apartir de la mayor cantidad de seguidores que tengan.
## Escribir un programa en python Pandas que pueda obtener esos usuarios

```python
jugadores = pd.read_csv('jugadores.csv')
jugadores
```

```python
# Obtenemos los valores maximos para cada pais, comparamos cantidad de seguidores de cada jugador con el maxim por pais
# Se pide ademas que la cantidad de seguidos no supere el 15% de los seguidores
jugadores['max_seg_pais'] = jugadores.groupby('pais',as_index = False)['cantidad_seguidores'].transform('max')
jugadores['cumplen'] = ((jugadores['cantidad_seguidores'])>(jugadores['max_seg_pais']*0.8)) & (jugadores['cantidad_seguidos']<jugadores['cantidad_seguidores']*0.15)
jugadores_que_cumplen = jugadores [(jugadores['cumplen']==True)]  
jugadores_que_cumplen
```

```python
#Ordenamos de mayor a menor por cant de seguidores
posibles_influencer = jugadores_que_cumplen[['id_usuario','nickname','pais','cantidad_seguidores']]
posibles_influencer.nlargest(5,'cantidad_seguidores')
```
