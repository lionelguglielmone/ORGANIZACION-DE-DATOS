```python
def get_trimestre(s):
    lista = s.split('-')
    mes = int(lista[1])
    trimestre=0
    if (mes==1) or (mes==2) or (mes==3):
        trimestre = 1
    if (mes==4) or (mes==5) or (mes==6):
        trimestre = 2
    if (mes==7) or (mes==8) or (mes==9):
        trimestre = 3
    if (mes==10) or (mes==11) or (mes==12):
        trimestre = 4
    return trimestre

def estado_del_caso(s):
    estado='NaN'
    if (s==1) or (s==3):
        estado = 'no_resuelto'
    else:
        estado = 'resuelto'
    return estado
```

```python
#aplicamos la funcion para definir el trimestre al que pertenece el delito
casos_short = casos[['fecha','estado_caso','categoria']]
casos_short['trimestre'] = casos_short['fecha'].apply(lambda x: get_trimestre(x))
#casos_short['estado'] = casos_short['estado_caso'].apply(lambda x: estado_del_caso(x))
```

```python
casos_short.head()
```

```python
casos_por_trimestre = casos_short[['trimestre','categoria']]
casos_por_trimestre['cantidad'] = 1
totales_por_trimestre = casos_por_trimestre.groupby(['trimestre','categoria'])['cantidad'].agg('count')
```

```python
totales_por_trimestre = totales_por_trimestre.reset_index()
totales_por_trimestre['trim-cat'] = (totales_por_trimestre['trimestre']).astype(str) + '-' + totales_por_trimestre['categoria']
totales_por_trimestre = totales_por_trimestre[['trim-cat','cantidad']]
totales_por_trimestre
```

```python
resueltos_por_trimestre = casos_short[casos_short['estado_caso']==2]
resueltos_por_trimestre = resueltos_por_trimestre[['trimestre','categoria']]
resueltos_por_trimestre['cantidad_resueltos'] = 1
resueltos_por_trimestre
```

```python
casos_resueltos_por_trimestre = resueltos_por_trimestre.groupby(['trimestre','categoria'])['cantidad_resueltos'].agg('count')
casos_resueltos_por_trimestre
```

```python
casos_resueltos_por_trimestre = casos_resueltos_por_trimestre.reset_index()
casos_resueltos_por_trimestre['trim-cat'] = (casos_resueltos_por_trimestre['trimestre']).astype(str) + '-' + casos_resueltos_por_trimestre['categoria']
casos_resueltos_por_trimestre = casos_resueltos_por_trimestre[['trim-cat','cantidad_resueltos']]
casos_resueltos_por_trimestre
```

```python
resueltos_vs_totales = pd.merge(casos_resueltos_por_trimestre,totales_por_trimestre, on = 'trim-cat')
resueltos_vs_totales
```

```python
resueltos_vs_totales['tasa'] = resueltos_vs_totales['cantidad_resueltos']/resueltos_vs_totales['cantidad']
resueltos_vs_totales_final = resueltos_vs_totales[['trim-cat','tasa']]
resueltos_vs_totales_final
```

```python
resueltos_vs_totales_final['diferencia'] = resueltos_vs_totales_final['tasa'].diff()
```
