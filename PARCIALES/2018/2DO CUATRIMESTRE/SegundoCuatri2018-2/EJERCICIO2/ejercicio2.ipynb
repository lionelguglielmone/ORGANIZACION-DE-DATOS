```python
products_short = (products[['product_id','category']])
products_short
```

```python
products_short = products_short[(products_short['category']=='men')]
products_short
```

```python
men_product_sale = products_short.join(sales_short)
men_product_sale
```

```python
men_product_sale ['promedio_mensual'] = men_product_sale.groupby('mes', as_index = False)['importe'].transform('mean')
men_product_sale
```

```python
men_product_sale ['precio_mayor_media'] = men_product_sale ['promedio_mensual']<men_product_sale ['importe']
men_product_sale
```

```python
expensive_men_products = men_product_sale[(men_product_sale['precio_mayor_media']==True)]
expensive_men_products = expensive_men_products[['product_id','importe','promedio_mensual']]
expensive_men_products
```
