```python
import pandas as pd
```

```python
reviews = pd.read_csv('ebooks-ratings.csv')
```

```python
books = pd.read_csv('good_reads.csv')
```

```python
reviews_by_book = reviews.groupby('book_id').agg({'rating':'mean'}).sort_values('rating',ascending = False)
```

```python
#para el punto B, primero hago merge de los data frames
books_and_reviews = pd.merge(reviews_by_book,books, on = 'book_id')
```

```python
books_and_reviews.drop(columns = 'book_name', inplace = True)
```
