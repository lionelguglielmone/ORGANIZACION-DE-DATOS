```python
eventos = pd.read_csv('eventos.csv')
```

```python
locacion = pd.read_csv('locacion.csv')
```

```python
#aunque los dataframes tiene distinto tamanio, se puede hacer merge si tienen alguna columna igual
locacion_evento = pd.merge(locacion,eventos,on='id_locacion')
```

```python
locacion_evento
```

```python
locacion_evento_short = locacion_evento[['nombre','capacidad','capacidad_extendida','sede','fecha','nombre_evento','cantidad_espectadores']]
locacion_evento_short
```

```python
locacion_evento_short['fecha_datetime'] = pd.to_datetime(locacion_evento_short['fecha'])
locacion_evento_short
```

```python
locacion_evento_short['dia_evento'] = locacion_evento_short['fecha_datetime'].dt.day
locacion_evento_short
```

```python
locacion_evento_extra_short = locacion_evento_short[['nombre','capacidad','capacidad_extendida'\
                                                     ,'cantidad_espectadores','dia_evento']]
locacion_evento_extra_short
```

```python
locacion_evento_grouped = locacion_evento_extra_short.groupby(['nombre','dia_evento']).agg({'cantidad_espectadores':'sum'})
locacion_evento_grouped
```

```python
locacion_evento_grouped_unstacked = locacion_evento_grouped.unstack()
locacion_evento_grouped_unstacked
```

```python
#una forma util de aplanar el multiindex de columnas
level_one = locacion_evento_grouped_unstacked.columns.get_level_values(0).astype(str)
level_two = locacion_evento_grouped_unstacked.columns.get_level_values(1).astype(str)
locacion_evento_grouped_unstacked.columns = level_one + level_two

locacion_evento_grouped_unstacked
```

```python
locacion_evento_14_y_15 = locacion_evento_grouped_unstacked[['cantidad_espectadores14','cantidad_espectadores16']]
locacion_evento_14_y_15
```

```python
locacion_evento_14_y_15.columns = [['14_octubre','16_octubre']]
locacion_evento_14_y_15
```

```python
locacion_evento_14_y_15.plot(kind='bar')
```
